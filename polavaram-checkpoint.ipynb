### Model Architecture

```python
class FCN(nn.Module):
    def __init__(self):
        super().__init__()
        #encoding layers
        self.encoder = nn.Sequential(
    nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3),
    nn.ReLU(inplace=True),
    nn.Conv2d(in_channels=32, out_channels=32, kernel_size=3),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=2),

    nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3),
    nn.ReLU(inplace=True),
    nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=2),

    nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3),
    nn.ReLU(inplace=True),
    nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=2)
)
        
        #decoding layers
        self.decoder = nn.Sequential(
    nn.Conv2d(in_channels=128, out_channels=1, kernel_size=1),
    nn.AdaptiveAvgPool2d(1)   # Global Average Pooling
)
        
    def forward(self, x : torch.Tensor) -> torch.Tensor:
        #passing the image data through the encoder and the decoder
        x=self.encoder(x)
        x=self.decoder(x)
        return x = torch.flatten(x, 1)
```

### Initializing the Train Loader and Test Loader

```python
class FireTensorDataset(Dataset):
    def __init__(self, root_dir, split="train", transform=None):
        self.root_dir = root_dir
        self.split = split
        self.transform = transforms.Compose([
    transforms.Resize((244, 244)),
    transforms.Lambda(lambda t: t[:3, :, :])  # force 3 channels
])

        # Collect samples
        self.samples = []
        fire_dir = os.path.join(root_dir, "fire_images", split)
        nonfire_dir = os.path.join(root_dir, "non_fire_images", split)

        # Fire = label 1
        for fname in os.listdir(fire_dir):
            if fname.endswith(".pt"):
                self.samples.append((os.path.join(fire_dir, fname), 1))

        # Non-fire = label 0
        for fname in os.listdir(nonfire_dir):
            if fname.endswith(".pt"):
                self.samples.append((os.path.join(nonfire_dir, fname), 0))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]

        # Load tensor directly
        tensor = torch.load(path)

        # Apply transforms (if given, e.g., normalization)
        if self.transform:
            tensor = self.transform(tensor)

        return tensor, label
```

```python
train_dataset = FireTensorDataset("firre\\fire_dataset", split="train",transform = transforms.Compose([
    transforms.Resize((244, 244)),
    transforms.Lambda(lambda t: t[:3, :, :])
]))
test_dataset  = FireTensorDataset("firre\\fire_dataset", split="test",transform = transforms.Compose([
    transforms.Resize((244, 244)),
    transforms.Lambda(lambda t: t[:3, :, :])
]))

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

### Initializing and Training the Model

```python
device="cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
torch.manual_seed(7203)
model=FCN()
model.state_dict()
```

```python
lossfn=BCEWithLogitsLoss()
optim=SGD(params=model.parameters(),lr=0.01)
```

```python
model.to(device)
training=[]
testing=[]
epochs=90
for epoch in range(epochs):
    train_loss=0
    test_loss=0
    model.train()
    torch.cuda.empty_cache()
    for batch,(image,label) in enumerate(train_loader):
        image,label=image.to(device),label.to(device)
        label=label.float()
        y_logits=model(image)
        loss=lossfn(y_logits,label.unsqueeze(1))
        train_loss+=loss
        optim.zero_grad()
        loss.backward()
        optim.step()
    model.eval()
    with torch.inference_mode():
        for batch,(image,label) in enumerate(test_loader):
            image,label=image.to(device),label.to(device)
            label=label.float()
            test_logits=model(image)
            test_loss+=lossfn(test_logits,label.unsqueeze(1)).item()
            break
    train_loss/=len(train_loader)
    test_loss/=len(test_loader)
    training.append(train_loss)
    testing.append(test_loss)
    print(f"Epoch:{epoch} | Train loss:{train_loss:.5f} | Test loss:{test_loss:.5f}")
```

```python
training=[x.cpu() for x in training]
testing=[x for x in testing]
plt.plot(range(epochs),training)
plt.plot(range(epochs),testing)
plt.xlabel("loss")
plt.ylabel("epochs")
plt.show()
```
